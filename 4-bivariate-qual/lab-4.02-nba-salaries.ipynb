## Exercise 2 - NBA Salaries

The file NBA.csv contains annual salaries for all NBA basketball players in the season (2008–2009), check whether there is independence between position and salary. To do this, first change any hyphenated position such as C-F to 
the first listed, in this case C. (Presumably, this is 
the player's primary position.)     
Then remove $ and . in Annual Salary and make Annual Salary categorical with four categories: the first is all salaries below the first quartile, the second is all salaries from the first quartile to the median, and so on.  
First make a plot

Results of the main calculations:
  - χ² ≈ 3.0344 
  - p-value = 0.8045
  - We do not reject H0


  image.png

```python
df = pd.read_csv(
    "https://raw.githubusercontent.com/HoGentTIN/dsai-labs/main/data/NBA.csv", sep=";"
)


df.head(10)
```

```python
# Data cleanup
# Only keep first position
df.Position = df.Position.str[0]

# Remove $ sign from Annual Salary
df["Annual Salary"] = df["Annual Salary"].str.replace("$", "")
# Remove dots
df["Annual Salary"] = df["Annual Salary"].str.replace(".", "")
df.head(10)
```

```python
# Put Annual Salary into categories
# convert Annual Salary to int instead of string
df["Annual Salary"] = df["Annual Salary"].astype(int)

percentiles = [0.0, 0.25, 0.5, 0.75, 1.0]
quantiles = df["Annual Salary"].quantile(percentiles)


def annual_salary_to_salary_category(salary):
    if salary < quantiles[0.25]:
        return 0
    if salary < quantiles[0.5]:
        return 1
    if salary < quantiles[0.75]:
        return 2
    return 3


df["Salary Category"] = df["Annual Salary"].map(annual_salary_to_salary_category)
df.head()
```

```python
age_brand = pd.crosstab(df["Position"], df["Salary Category"])
age_brand.plot(kind="barh", stacked=True)
```

```python
# Chi-squared test for independence
chi2, p, df, expected = stats.chi2_contingency(age_brand)

print("Chi-squared       : %.4f" % chi2)
print("Degrees of freedom: %d" % df)
print("P-value           : %.4f" % p)
```
